```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.optimize import curve_fit
%config InlineBackend.figure_format='svg'
```

```python
I = np.array([.00706, .00514, .00404, .003325, .002828, .002433])
V = np.array([3.53, 5.14, 6.06, 6.65, 7.07, 7.38])
yerr = (.01*V)
plt.errorbar(I, V, yerr, fmt='.')

def mod(I, R0, V0):
    return(V0 + I*R0)
init_guess = [100, 1]

fit = scipy.optimize.curve_fit(mod, I, V, sigma = yerr, p0 = init_guess, absolute_sigma = True)
ans, cov = fit
fit_R0, fit_V0 = ans

z = np.linspace(0.0023, 0.0071)
plt.plot(z, mod(z, fit_R0, fit_V0), label = 'mod')
plt.title('Current vs Voltage')
plt.xlabel('Voltage(V)')
plt.ylabel('Current(A)')
plt.legend()
plt.figtext(0.2, 0.8, "slope(resistance): ")

print(fit_R0, fit_V0)
```
